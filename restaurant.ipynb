LABEL ENCODING FOR BINARY CATEGORIES

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
df['Switch to order menu'] = labelencoder.fit_transform(df['Switch to order menu'])
df['Has Table booking'] = labelencoder.fit_transform(df['Has Table booking'])
df['Has Online delivery'] = labelencoder.fit_transform(df['Has Online delivery'])
df['Is delivering now'] = labelencoder.fit_transform(df['Is delivering now'])
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
df[['Country Code','Longitude','Latitude','Average Cost for two']] = minmax_scaler.fit_transform(df[['Country Code','Longitude','Latitude','Average Cost for two']])
df.head()
```

```python
df['Rating color'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinalencoder = OrdinalEncoder(categories=[['Orange', 'White', 'Yellow','Green','Dark Green','Red']])
df['Rating color'] = ordinalencoder.fit_transform(df[['Rating color']])
df.head()
```

```python
df['Rating text'].value_counts()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
ordinalencoder = OrdinalEncoder(categories=[['Not rated','Poor','Average','Good','Very Good','Excellent']])
df['Rating text'] = ordinalencoder.fit_transform(df[['Rating text']])
df.head()
```

```python
df.info()
```

```python
df['Votes'].value_counts()
```

```python
df_dummies = pd.get_dummies(df, columns=['Restaurant Name', 'City','Address','Locality','Locality Verbose','Cuisines','Currency'], drop_first=True)
df_dummies.head()
```

```python
df_numeric = df_dummies *1
df_numeric
```

```python
df_numeric['Currency_Dollar($)'].value_counts()
```

```python
x = df_numeric.drop("Aggregate rating",axis=1)
y = df_numeric['Aggregate rating']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(random_state=42)
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python

feature_importance = pd.DataFrame(model.feature_importances_, index=x.columns, columns=['Importance']).sort_values(by='Importance', ascending=False)
feature_importance
```
